Load Up Dataset

```python
data = pd.read_csv("/kaggle/input/unemployment-dataset/unemployment analysis.csv")
names= data["Country Name"].to_list()
```

```python
free = data.transpose().copy()
free = free[2:]
free.columns = names
```

```python
worst_5 = data.sort_values(by="2021", ascending=False).head(5)
best_5 = data.sort_values(by="2021", ascending=True).head(5)
bottom_5_names = worst_5["Country Name"].to_list()
top_5_name = best_5["Country Name"].to_list()
```

```python
best_5 = best_5.transpose()[2:]
best_5.columns = top_5_name
best_5
```

# Best 5 Countries by Employment Rate 2021

```python
plt.figure(figsize=(10,10))
plt.plot(best_5.index, best_5.Qatar, linewidth=3, alpha=0.8, label="Qatar")
plt.plot(best_5.index, best_5.Cambodia, linewidth=3,alpha=0.8, label="Cambodia")
plt.plot(best_5.index, best_5.Niger, linewidth=3,alpha=0.8, label="Niger")
plt.plot(best_5.index, best_5["Solomon Islands"], linewidth=3,alpha=0.8, label="Solomon Isles")
plt.plot(best_5.index, best_5["Lao PDR"], linewidth=3,alpha=0.8, label="Lao")

plt.grid(axis="both", linestyle="--", alpha=0.6)
plt.legend()
plt.title("Best 5 countries for Employment Rates")
sns.despine()
plt.ylabel("Unemployment as a %")
plt.xticks(rotation=45);
```

```python
worst_5 = worst_5.transpose()[2:]
worst_5.columns = bottom_5_names
worst_5
```

# Highest 5 Countries Unemployment by 2021

```python
plt.figure(figsize=(10,10))
plt.plot(worst_5.index, worst_5["South Africa"], linewidth=3, alpha=0.8, label="South Africa")
plt.plot(worst_5.index, worst_5.Djibouti, linewidth=3,alpha=0.8, label="Djibouti")
plt.plot(worst_5.index, worst_5.Eswatini, linewidth=3,alpha=0.8, label="Eswatini")
plt.plot(worst_5.index, worst_5["West Bank and Gaza"], linewidth=3,alpha=0.8, label="West Bank and Gaza")
plt.plot(worst_5.index, worst_5["Botswana"], linewidth=3,alpha=0.8, label="Botswana")

plt.grid(axis="both", linestyle="--", alpha=0.6)
plt.legend()
plt.title("Worst 5 countries for Employment Rates")
sns.despine()
plt.ylabel("Unemployment as a %")
plt.xticks(rotation=45);
```

Creating new column with unemployment rates as a percentage change from 1991-2021, sorting the values from highest change to lowest.

```python
data["Change"] = ((data["1991"]-data["2021"])/data["1991"])*100
data = data.sort_values(by="Change", ascending=False)
data.head(10).reset_index(drop=True)
improved = data["Country Name"].to_list()
data = data.head(10).transpose()[2:]
data.columns = improved[:10]
```

```python
data.head(5)
```

# Most imporved Countries By Employment Rates from 1991-2021

```python
plt.figure(figsize=(10,10))
plt.plot(data.index[:-1], data["Trinidad and Tobago"][:-1], label="Trinidad and Tobago")
plt.plot(data.index[:-1], data["Poland"][:-1], label="Poland")
plt.plot(data.index[:-1], data["Qatar"][:-1], label="Qatar")
plt.plot(data.index[:-1], data["Cuba"][:-1], label="Cuba")
plt.plot(data.index[:-1], data["Sri Lanka"][:-1], label="Sri Lanka")
plt.plot(data.index[:-1], data["Israel"][:-1], label="Isreal")
plt.plot(data.index[:-1], data["New Zealand"][:-1], label="New Zealand")
plt.plot(data.index[:-1], data["Ireland"][:-1], label="Ireland")
plt.plot(data.index[:-1], data["Barbados"][:-1], label="Barbados")
plt.plot(data.index[:-1], data["Hungary"][:-1], label="Hungary")

plt.legend()
plt.ylabel("Unemployment as a %")
sns.despine()
plt.title("Most Improved Countries")
plt.grid(axis="both", linestyle="--", alpha=0.6)
plt.xticks(rotation=45);
```
